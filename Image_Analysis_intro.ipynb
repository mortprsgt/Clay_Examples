```python
%run NBCONFIG.ipynb

from skimage.data import coins
from skimage.filter import threshold_adaptive
from skimage.segmentation import clear_border
from pyparty.plots import showim

thresh_ad = clear_border(threshold_adaptive(coins(), 131))
showim(thresh_ad)
```

```python
from skimage.morphology import label

labels = label(thresh_ad, neighbors=8)
showim(labels, 'spectral');
```

```python
np.unique(labels)
mask = (labels > 200) & (labels < 250)
showim(mask, 'gray')
```

```python
from pyparty import Canvas
from skimage import data

coins = data.coins()

showim(coins);
```

```python
c = Canvas(background = coins)
c.show(grid=True)
```

```python
c.add('circle', radius=100, color='green')
print c
c.show;
```

```python
def red_or_blue(p):
    if p.area >2000:
        p.color = 'red'
    else:
        p.color = (0,0,1) #blue
    return p
cfinal = c2.pmap(red_or_blue)
cfinal.show()
```

```python
c2.particles = c2[c2.area > 2000]
c3 = c2.pmap(red_or_blue)
c3.patchshow()
```
